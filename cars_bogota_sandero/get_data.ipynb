```python
def get_price(sout):
    price = soup.find('span', attrs={'class':'price-tag-fraction'}).text
    
    return {'price': price}
```

```python
def get_data_sheet(soup):
    car_data = dict()
    
    data_sheet = soup.find_all('li', attrs={'class':'specs-item'})
    for item in data_sheet:
        key = item.find('strong').text
        value = item.find('span').text
        car_data.update({key: value})
    
    return car_data
```

```python
def get_additional(soup):
    car_data = dict()
    
    additional = soup.find('div', attrs={'class': 'description-content-secondary-group'})
    additional_items = additional.find_all('li')
    for item in additional_items:
        try:
            key, value = delete_dirty_chars(item.text).split(':')
            car_data.update({key: value})
        except Exception as e:
            pass
    return car_data
```

```python
def delete_dirty_chars(string):
    return string.replace(' ', '').replace('\n', '').replace('\r', '').replace('\t', '')
```

```python
all_data = []
for index, url in enumerate(list_urls):
    car_data = dict()
    
    r = requests.get(url)
    if r.status_code == 404:
        continue
    soup = BeautifulSoup(r.text, 'html.parser')
    
    car_data.update(get_price(soup))
    car_data.update(get_data_sheet(soup))
    car_data.update(get_additional(soup))
    
    all_data.append(car_data)
    print('--------------------------------------------------------------------------------')
    print(f'{index} - {url}')
    print(car_data)
    print('--------------------------------------------------------------------------------')
```

```python
all_data_table = []

for car in all_data:
    price = car.get('price')
    price = price if price else 'nan'
    
    brand = car.get('Marca')
    brand = brand if brand else 'nan'
    
    model = car.get('Modelo')
    model = model if model else 'nan'
    
    version = car.get('Versión')
    version = version if version else 'nan'
    
    year = car.get('Año')
    year = year if year else 'nan'
    
    milleage = car.get('Kilómetros')
    milleage = milleage if milleage else 'nan'
    
    doors = car.get('Puertas')
    doors = doors if doors else 'nan'
    
    fuel = car.get('Tipo de combustible')
    fuel = fuel if fuel else 'nan'
    
    shield = car.get('Placa')
    shield = shield if shield else 'nan'
    
    color = car.get('Color')
    color = color if color else 'nan'
    
    transmission = car.get('Transmisión')
    transmission = transmission if color else 'nan'
    
    direction = car.get('Dirección')
    direction = direction if direction else 'nan'
    
    traction = car.get('Controldetracción')
    traction = traction if traction else 'nan'
    
    only_owner = car.get('Únicodueño')
    only_owner = only_owner if only_owner else 'nan'
    
    motor = car.get('Motor')
    motor = motor if motor else 'nan'
    
    cylinders = car.get('Cilindros')
    cylinders = cylinders if cylinders else 'nan'
    
    car_info = (price, brand, model, version, year, milleage, doors, fuel, shield,
               color, transmission, direction, traction, only_owner, motor, cylinders)
    all_data_table.append(car_info)
```

```python
columns = ['Price', 'Brand', 'Model', 'Version', 'Year', 'Milleage', 'Doors', 'Fuel', 'Shield', 
           'Color', 'Transmision', 'Direction', 'Traction', 'Only owner', 'Motor', 'Cylinders']

data_frame = pd.DataFrame(all_data_table, columns=columns)
```

```python
data_frame.to_csv('renault_cars.csv', index=False, encoding='utf-8')
```
