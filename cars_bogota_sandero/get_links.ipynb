```python
import requests
from bs4 import BeautifulSoup
```

```python
place_url = 'https://carros.tucarro.com.co/renault/sandero/bogota-dc'
request_url = place_url
all_links = []
while True:
    r = requests.get(request_url)
    if r.status_code == 404:
        break
    soup = BeautifulSoup(r.text, 'html.parser')
    results = soup.find_all('li', attrs={'class':'results-item highlighted article grid'})
    links = [result.find('a', attrs={'class': 'item-link item__js-link'})['href'] for result in results]
    all_links.extend(links)
    request_url = place_url + f'/_Desde_{len(all_links)}'
    print(request_url)
```
